```python
pd.Categorical(data_19["Company type"])
```

```python
e=[]
f=[]
g=[]
h=[]
k=[]
l=[]
m=[]
for i in range (len(pd.Categorical(data_19["Company type"]))):
    if (data_19["Company type"][i] == 'Bank'):
        e.append(i)
    if (data_19["Company type"][i] == 'Bodyshop / Outsource'):
        f.append(i)
    if (data_19["Company type"][i] == 'Consulting / Agency'):
        g.append(i)
    if (data_19["Company type"][i] == 'Outsource'):
        h.append(i)
    if (data_19["Company type"][i] == 'Product'):
        k.append(i)
    if (data_19["Company type"][i] == 'Startup'):
        l.append(i)
    if (data_19["Company type"][i] == 'University'):
        m.append(i)
print("Number of Bank :",len(e))
print("Number of Bodyshop / Outsource :",len(f))
print("Number of Consulting / Agency :",len(g))
print("Number of Outsource :",len(h))
print("Number of Product :",len(k))
print("Number of Startup :",len(l))
print("Number of University :",len(m))
if (len(g) > len(l)):
    print("Consulting / Agency companies are more popluar than Startups")
else:
    print("Startups are more popluar than Consulting / Agency companies ")
```

```python
import matplotlib.pyplot as plt  
data=[6,30,117,1,619,181,6]
explode = (0.0, 0.0, 0.5, 0.0, 0.0, 0.5 , 0.0)
labels=['Bank', 'Bodyshop / Outsource', 'Consulting / Agency', 'Outsource', 'Product', 'Startup', 'University']
plt.pie(data, explode=explode,autopct='%1.2f%%',shadow=True, radius=2)
plt.legend(labels, loc=10)
plt.show()
```
